```python
import pandas as pd
```

```python
file1 = pd.read_csv("color_web_safe.csv")
file1.head()#data showing
```

```python
file1.isnull()#finding null value
```

```python
#finding sum of null value
file1.isnull().sum().sum()
```

```python
#fill value with zero
file1.fillna(0)
file1.head()
```

```python
#filling value with back value
file1.fillna(method="bfill")
file1.head()
```
