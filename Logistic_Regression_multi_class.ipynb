```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
veg_table_file = pd.read_csv("avocado_ripeness_dataset.csv")
veg_table_file.head()
```

Finding Null Values

```python
veg_table_file.isnull().sum()
```

Encode Target Column
hard → 0  
breaking → 1  
ripe → 2  
pre-conditioned → 3


```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
veg_table_file['ripness_encoder'] = encoder.fit_transform(veg_table_file['ripeness'])
print(veg_table_file[['ripeness', 'ripness_encoder']])
```

Define Features (X) and Target (y)


```python
x = veg_table_file.drop('ripness_encoder' ,axis= 1)
y = veg_table_file['ripness_encoder']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model  = LogisticRegression()

```

```python
X_train = pd.get_dummies(X_train, columns=['color_category'], drop_first=True)
X_test = pd.get_dummies(X_test, columns=['color_category'], drop_first=True)
```

```python
print(X_train)

```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
X_train['ripeness']= label_encoder.fit_transform(X_train['ripeness'])
X_test['ripeness']= label_encoder.fit_transform(X_test['ripeness'])


X_train['ripeness'].unique()
```

```python
print(Y_train)
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# Accuracy (how many correct)
print("Accuracy:", accuracy_score(Y_test, y_pred))

# Confusion Matrix
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred))

# Detailed classification report
print("Classification Report:\n", classification_report(Y_test, y_pred))
```
